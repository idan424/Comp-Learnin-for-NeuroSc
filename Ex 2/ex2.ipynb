# Q1


a. Train a Linear Regression model which predicts "Visc" from "BMI", "Waist circumference" and WHtR ("Waist circumference"/"Height_value-1").

b. Report the root mean squared prediction error (RMSE) on the test set.

c. Report the estimated coefficient's values.

d. Plot a scatter plot of the predicted target variable v.s. the actual target variable based on
the test set.

```
target = "Visc"
data = df[["Visc", "BMI", "Waist circumference"]].copy()
data["WHtR"] = df["Waist circumference"]/df["Height_value-1"]

y = data.pop(target)
X = data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

> a.

```
model = LinearRegression()
model.fit(X_train, y_train)
```

> b.

```
y_pred = model.predict(X_test)
print(f"the RMSE is: {MSE(y_test, y_pred, squared=False)}")
```

> c.

```
print(f"beta 0 is: {model.intercept_}")
print(f"beta[1-3] are: {model.coef_}")
```

> d.

```
_ = sns.scatterplot(x=y_pred, y=y_test)
plt.title("predicted v.s. target")
plt.xlabel("y_pred")
plt.ylabel("y_test")
```

# Q2

Repeat question 1 - where the predictors are transformed to categorical variables based on their quartiles (with levels: 0-25th percentile, 25-50th percentile, 50-75th percentile, 75-100th percentile).

```
target = "Visc"
data = df[["Visc", "BMI", "Waist circumference"]]
data["WHtR"] = df["Waist circumference"]/df["Height_value-1"]

y = data.pop(target)
X = data
```

> This is the new section for Q2:


```
for col in X.columns:
    X[col] = pd.qcut(x=X[col], q=4, labels=False)
    # get the dummies
    dummies = pd.get_dummies(X[col], prefix=col, drop_first=True)
    # dummie column names
    dcol = dummies.columns
    # add new dummy encoding
    X[dcol] = dummies
    X.pop(col)
X
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

> a.

```
model = LinearRegression()
model.fit(X_train, y_train)
```

> b.

```
y_pred = model.predict(X_test)
print(f"the RMSE is: {MSE(y_test, y_pred, squared=False)}")
```

> c.

```
print(f"beta0 is: {model.intercept_}")
print(f"beta[1-3] are: {model.coef_}")
```

> d.

```
_ = sns.scatterplot(x=y_pred, y=y_test)
plt.title("predicted v.s. target")
plt.xlabel("y_pred")
plt.ylabel("y_test")
```

# Q4

```
target = "Visc"
data = df[["Visc", "BMI", "Waist circumference"]]
data["WHtR"] = df["Waist circumference"]/df["Height_value-1"]

X = data.copy()
y = data.copy().pop(target)
```

> a.


```
y = (y > y.median()).astype("int8")
```

> b.

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

> c.

```
roc_auc_score(y_test, y_pred)
```

> d.

```
conf = confusion_matrix(y_test, y_pred)
print(f"confusion matrix:\n{conf}")

tn, fp, fn, tp = conf.ravel()

specificity = tn / (tn + fp)
sensitivity = tp / (tp + fn)

print(f"\nSpecificity: {specificity},Sensitivity: {sensitivity}")
```
